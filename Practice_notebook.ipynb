# Reading Group Properties

```python

#----------Here we set the path to the simulation-------------
path_to_output='/orange/lblecha/aklantbhowmick/GAS_BASED_SEED_MODEL_UNIFORM_RUNS/L12p5n512/' # this is the folder containing the simulation run
run='/AREPO/' # name of the simulation runs
output='output_ratio10_SFMFGM5_seed5.00_bFOF/'
basePath=path_to_output+run+output



#Here we set the 'desired redshift', 
#the code will automatically select the closest available redshift snapshot, 
#and set that as 'output_redshift'

desired_redshift=0

#Reading the total halo mass
GroupMass,output_redshift=arepo_package.get_group_property(basePath,'GroupMass',desired_redshift,postprocessed=1)
#Reading the total halo mass for each type, dark matter=0, gas=1, star=4, black hole=5
GroupMassType,output_redshift=arepo_package.get_group_property(basePath,'GroupMassType',desired_redshift,postprocessed=1)
#Reading the total black hole mass of the halo, different particle type 5
GroupBHMass,output_redshift=arepo_package.get_group_property(basePath,'GroupBHMass',desired_redshift,postprocessed=1)


print("Desired_redshift=",desired_redshift)
print("Output_redshift=",output_redshift)

GroupSM=GroupMassType[:,4]
```

```python
tick_size=40
label_font_size=50

f,axx=plt.subplots(1,1,figsize=(16,14),sharey=True,sharex=True)
axx.errorbar(GroupMass*1e10,GroupSM*1e10,marker='o',linewidth=0)
axx.set_xscale('log')
axx.set_yscale('log')
axx.tick_params(labelsize=tick_size)

axx.set_xlabel('$M_{total} (M_{\odot}/h)$ halos',fontsize=label_font_size)
axx.set_ylabel('$M_* (M_{\odot}/h)$ halos',fontsize=label_font_size)



```
